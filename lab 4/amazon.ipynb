```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def get_cleaned_textdata(sentence):
    # Remove HTML tags
    modified_sentence = re.sub(r'<.*?>', ' ', sentence)
    
    # Remove punctuation and special characters
    modified_sentence = ''.join([i if i not in string.punctuation else ' ' for i in modified_sentence])
    
    # Remove digits
    modified_sentence = re.sub(r'\d', ' ', modified_sentence)
    
    # Convert to lowercase
    modified_sentence = modified_sentence.lower()
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    modified_sentence = ' '.join([word for word in modified_sentence.split() if word not in stop_words])
    
    # Perform stemming
    stemmer = PorterStemmer()
    modified_sentence = ' '.join([stemmer.stem(word) for word in modified_sentence.split()])
    
    # Perform lemmatization
    lemmatizer = WordNetLemmatizer()
    modified_sentence = ' '.join([lemmatizer.lemmatize(word) for word in modified_sentence.split()])
    
    return modified_sentence

# Example usage:
sentence = "This is a test sentence, with some <html>tag</html> and digits like 123."
cleaned_sentence = get_cleaned_textdata(sentence)
print(cleaned_sentence)
```

```python
import pandas as pd
import re
import string

def get_cleaned_textdata(sentence):
    # Remove HTML tags
    modified_sentence = re.sub(r'<.*?>', '', sentence)
    
    # Remove punctuation
    modified_sentence = ''.join([i if i not in string.punctuation else '' for i in modified_sentence])
    
    # Remove digits
    modified_sentence = re.sub(r'\d', '', modified_sentence)
    
    # Remove extra whitespaces
    modified_sentence = re.sub(r'\s+', ' ', modified_sentence)
    
    # Convert to lowercase
    modified_sentence = modified_sentence.lower()
    
    return modified_sentence

# Read the data from the Excel file
data = pd.read_csv("k_Reviews.csv")

# Remove duplicates based on the 'Text' column
text_column = data['Text'].drop_duplicates()

# Apply the cleaning function to the 'Text' column after removing duplicates
data['modified_sentence'] = text_column.apply(get_cleaned_textdata)

# Display the modified text column
print(data['modified_sentence'])
```
